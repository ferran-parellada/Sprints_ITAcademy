# Donar format a text amb Markdown

## Resum de les regles bàsiques:


El format es defineix amb etiquetes específiques per a cada cas:

+ El parragraf es forma amb dos return.
+ El salt de línia amb dos espais seguits.
+ Les cites amb > al principi del text


+ Text en itàlica : \*...\* o \_...\_ 
+ Text en negreta : \*\*...\*\* o \_\_...\_\_


+ Títols : \# 
+ Llistes desordenades : \* + espai o \_ + espai o \+ + espai. Per als subnivells es pot precedir d'un espai o el tabulador.
+ Llistes ordenades :  1., 2., etc.


+ Codi, forma 1 :  entre dos \`\` (\`\`codi\`\`). Equival al \<code\> de l'html.
+ Codi, forma 2 : precedit de quatre espais en cada línia. Equival al \<pre\> de l'html.
+ Blocs de codi : entre dues línies formades per tres ~ (~~~)


+ Enllaços ocultant l'adreça : el nom entre [] i l'adreça de l'enllaç entre ()
+ Enllaços mostrant l'adreça : adreça continguda entre \< i \> 
+ Imatges : \![text alternatiu](ruta de la imatge)

+ \*\*\* sense text : Dibuixa una línia horitzontal


## Donar estil al text

* Itàlica : contingut entre asteriscs o subguions : \*text\* o \_text\_
    * _Estic entre subguions_. Altre exemple: sóc un text normal *i jo en itàlica*.
    
    
    
* Negreta : contingut entre dobles asteriscs o subguions: \*\*text\*\* o \_\_text\_\_
    * __Estic entre dobles subguions__. Jo sóc un text normal **i jo estic entre dobles asteriscs**.



## Títols

* Els títols van precedits del símbol #.
* N'hi ha fins a sis nivells, en progressió descendent (equivalents a l'etiqueta h1, h2, etc., de l'html).
    * Exemple: # Això serà un títol de nivell 1
* Es poden 'escapar' amb la contrabarra \ : 

Mostres:
# Títol nivell 1
## Títol nivell 2
### Títol nivell 3
#### Títol nivell 4
##### Títol nivell 5
###### Títol nivell 6

## Llistes desordenades
\* + espai + text

* Nivell 1 de la llista
 * Nivell 2 de la llista

## Llistes ordenades
1. Nivell 1
2. Nivell 1

    2.1. Nivell 2
    
    2.2. Nivell 2
