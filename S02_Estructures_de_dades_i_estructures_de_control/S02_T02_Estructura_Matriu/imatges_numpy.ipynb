### Exercici 3

Crea una matriu bidimensional amb una forma de 5 x 5. Extreu el valor màxim de la matriu, i els valors màxims de cadascun dels seus eixos.

```python
array5x5 = np.random.randint(100, size=(5, 5))

print(array5x5)

maxInArray = np.max(array5x5)

print("El valor màxim en l'array és",maxInArray)

maxInRows = list(map(np.max, array5x5))

print("Els valors màxims de cada filera són:", maxInRows)

maxInColumns = list(map(np.max, np.transpose(array5x5)))

print("Els valors màxims de cada columna són:", maxInColumns)
```

### Exercici 4

Mostreu-me amb exemples de diferents matrius, la regla fonamental de Broadcasting que diu : "les matrius es poden transmetre / broadcast si les seves dimensions coincideixen o si una de les matrius té una mida d'1".

```python
# Referència, la documentació oficial de numpy: https://numpy.org/doc/stable/user/basics.broadcasting.html
from numpy.random import randn

# Array (3, 3, 3)
arr1 = np.array(randn(3, 3, 3) * 10, dtype= np.int64)
print("arr1:", arr1, sep="\n")

# Array (1, 1, 3). Per tant, té només una filera i una columna, però tres elements en la darrera dimensió.
arr2 = np.array([[[0, 1, 0]]])
print("=" * 25)
print("arr2:", arr2, sep="\n")

# Quan s'opera amb arr1 i arr2 es produeix el broadcasting (propagació),
# amb la multiplicació del primer element d'array2 per cada primer element d'arr1,
# el segon d'arr2 pels segons elements d'arr1, etc.
print("=" * 25)
print("arr1 * arr2:", arr1 * arr2, sep="\n")

# Passa el mateix amb un array de mida 1
arr3 = np.array([0, 1, 0])
print("=" * 25)
print("arr3:", arr3, sep="\n")
print("arr1 * arr3:", arr1 * arr3, sep="\n")
```

Manipulació d’imatges amb Matplotlib.

Carregareu qualsevol imatge (jpg, png ..) amb Matplotlib. adoneu-vos que les imatges RGB (Red, Green, Blue) són realment només amplades × alçades × 3 matrius (tres canals Vermell, Verd i Blau), una per cada color de nombres enters int8,

manipuleu aquests bytes i torneu a utilitzar Matplotlib per desar la imatge modificada un cop hàgiu acabat.

Ajuda:Importeu, import matplotlib.image as mpimg. estudieu el metodde mpimg.imread(()

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

jpgImage1 = 'nefertiti.jpg'

# Utilitzaré mpimg.imread(), però també serviria plt.imread()
arrImage1 = mpimg.imread(jpgImage1, format='jpg')

print(arrImage1)
print("Dimensió de l'array:", arrImage1.ndim)
print("Tipus dades:", arrImage1.dtype) 

# L'array representa la imatge de manera que la primera dimensió és l'alçada i la segona l'amplada
# La tercera dimensió correspon als canals RGB (Red-Green-Blue, i per tant és tres) o RGBA (incoporant l'Alpha, i serien 4)
# El valor del píxels de cada canal pot anar entre 0 (completament negre) i 255 (completament blanc)
# Entremig estan els matissos de gris
print("Forma de l'array:", arrImage1.shape, "->", arrImage1.shape[0], "x", arrImage1.shape[1], "són les mides de la imatge en píxels")

# Representació amb eixos
plt.imshow(arrImage1)
plt.title("Imatge original")
plt.colorbar()
```

```python
# Representació sense eixos
plt.axis('off')
plt.imshow(arrImage1)
plt.title("Imatge original sense eixos")
```

```python
# Els eixos ajuden si es vol el detall d'alguna part
detallImatge = arrImage1[400:800, 125:600, :]
plt.imshow(detallImatge)
plt.title("Detall de la imatge")
plt.colorbar()
```

```python
# Sel·leció només del canal verd
plt.imshow(detallImatge[:, :, 1])
plt.title("Només el canal verd")

# Guardar la imatge
mpimg.imsave("Nefertiti_nomes_canal_verd.jpg", arrImage1[:, :, 1])
```
